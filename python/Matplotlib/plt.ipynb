```python
# -*- coding: utf-8 -*-

import numpy as np
import matplotlib.pyplot as plt

% matplotlib inline
```

```python
# 散点图
plt.scatter(np.random.randn(100), np.random.randn(100), marker='+', c=['red', 'green', 'purple', 'yellow'], linewidths=2)
# 横轴
plt.xlabel('x')
# 纵轴
plt.ylabel('y')
# 标题
plt.title('Scatter Plot')
```

```python
# 子图
nums = 20
basepos = 420
# figsize 指定图形的宽度和高度
plt.figure(figsize=(10, 12))
plt.subplot(basepos + 1)
plt.plot(np.random.randn(nums), np.random.randn(nums), 'g-')
plt.subplot(basepos + 2)
plt.plot(np.random.randn(nums), np.random.randn(nums), 'g--')
plt.subplot(basepos + 3)
plt.plot(np.random.randn(nums), np.random.randn(nums), 'g.')
plt.subplot(basepos + 4)
plt.plot(np.random.randn(nums), np.random.randn(nums), 'go')
plt.subplot(basepos + 5)
plt.plot(np.random.randn(nums), np.random.randn(nums), 'gv')
plt.subplot(basepos + 6)
plt.plot(np.random.randn(nums), np.random.randn(nums), 'g^')
plt.subplot(basepos + 7)
plt.plot(np.random.randn(nums), np.random.randn(nums), 'gs')
plt.subplot(basepos + 8)
# linewidth 线条宽度
plt.plot(np.random.randn(nums), np.random.randn(nums), 'g+', linewidth=3)
```

```python
nums = 50
# 混合图
plt.plot(np.random.randn(nums), np.random.randn(nums), 'r-', np.random.randn(nums), np.random.randn(nums), 'go')
# 可视区域
plt.axis([0, 2, 0, 2])
```

```python
# 模拟曲线图
x = np.linspace(0, 10, 1000)
plt.figure(figsize=(8, 4))
plt.plot(x, np.sin(x), 'r-')
plt.plot(x, np.cos(x), 'g--')
```

```python
# 直方图
plt.hist(np.random.randn(200))
```

```python
# 箱线图
plt.boxplot(np.random.randn(100))
```

```python
# 饼装图
plt.pie([50, 30, 20], labels=['Java', 'C', 'Python'])
```

```python
# 柱状图
x = np.arange(1, 11)
y = np.array(np.random.randn(10) * 100)
plt.bar(x, y, width=0.8)
```

```python
# 横线
plt.axhline(5, xmin=0, xmax=0.75, linewidth=2, color='r')
plt.axis([0, 10, 0, 10])
# 竖线
plt.axvline(5, ymin=0.25, ymax=1, linewidth=2, color='g')
plt.axis([0, 10, 0, 10])
```
